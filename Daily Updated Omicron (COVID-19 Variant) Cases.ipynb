```python
avg_total_deaths = df.groupby("continent")["total_deaths"].sum()
print(avg_total_deaths)
```

```python
avg_total_deaths.plot(x="Million", y="Continent", kind="bar", title = "Total Deaths by Continent")
plt.show()
```

```python
avg_total_new_cases = df.groupby("continent")["new_cases"].sum()
print(avg_total_new_cases)
```

```python
avg_total_new_cases.plot(x="Million", y="location", kind="bar", title = "Total New Cases by Continent")
plt.show()
```
