```python
import pandas as pd
import matplotlib.pyplot as plt
brics = pd.read_csv("H:\Python\owid-covid-data.csv", index_col = 0)
brics
brics["total_deaths"].hist()
plt.show()
```

```python
avg_total_deaths = brics.groupby("continent")["total_deaths"].sum()
print(avg_total_deaths)
```

```python
avg_total_deaths.plot(x="Million", y="Continent", kind="bar", title = "Total Deaths by Continents")
plt.show()
```

```python
avg_total_new_cases = brics.groupby("location")["new_cases"].sum()
print(avg_total_new_cases)
```

```python
avg_total_new_cases.plot(x="Million", y="location", kind="bar", title = "Total New Cases by Country")
plt.show()
```
